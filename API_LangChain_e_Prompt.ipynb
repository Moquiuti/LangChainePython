```python
from dotenv import load_dotenv
from openai import OpenAI
import os

load_dotenv()

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

dias = 5
numero_criancas = 2
atividade = "atividades ao ar livre"

prompt = f"""
Crie um roteiro de viagem para uma família.
Destino com duração de {dias} dias,
com {numero_criancas} crianças,
focado em {atividade}.
"""

resposta = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "Você é um assistente especializado em criar roteiros de viagem."},
        {"role": "user", "content": prompt}
    ]
)

print(resposta.choices[0].message.content)
```
